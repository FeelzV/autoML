```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
from utils import load
blobs_data_train = h2o.import_file("datasets/blobs_train.csv")
blobs_data_test = h2o.import_file("datasets/blobs_test.csv")
blobs_data_train['label'] = blobs_data_train['label'].asfactor()
blobs_data_test['label'] = blobs_data_test['label'].asfactor()
blobs_predictor = H2OAutoML(max_models=10, seed=1)
blobs_predictor.train(y='label', training_frame=blobs_data_train)
```

```python
blobs_predictor.leader.model_performance(blobs_data_test)
blobs_predictor.leader.explain(blobs_data_test)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
import pandas as pd
vectorizer = CountVectorizer()
df = pd.read_csv("datasets/emotion.csv")
vectorizer.fit(df['text'])
vector = vectorizer.transform(df['text'])
df['text'] = list(vector.toarray())
emotion_data_train, emotion_data_test = train_test_split(df, test_size=0.2, random_state=0)
emotion_data_train = h2o.H2OFrame(emotion_data_train)
emotion_data_test = h2o.H2OFrame(emotion_data_test)
emotion_data_train['label'] = emotion_data_train['label'].asfactor()
emotion_data_test['label'] = emotion_data_test['label'].asfactor()
print(emotion_data_train)
```

```python
emotion_predictor = H2OAutoML(max_models=10, seed=1)
emotion_predictor.train(y='label', training_frame=emotion_data_train)
```

```python
emotion_predictor.leader.model_performance(emotion_data_test)
```
